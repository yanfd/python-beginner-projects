Model link: https://huggingface.co/warp-ai/wuerstchen

To run the model, Open the .ipynb file in google colab and change the notebook settins to T4 GPU.

Later change the caption to which ever photo you want to generate.

```python
import torch
from diffusers import AutoPipelineForText2Image

device = "cuda"
dtype = torch.float16

pipeline =  AutoPipelineForText2Image.from_pretrained(
    "warp-diffusion/wuerstchen", torch_dtype=dtype
).to(device)

caption = "Anthropomorphic cat dressed as a fire fighter"

output = pipeline(
    prompt=caption,
    height=1024,
    width=1024,
    prior_guidance_scale=4.0,
    decoder_guidance_scale=0.0,
).images
```
